```python
df.set_index('DateTime',inplace=True)
df.head()
```

```python
from sklearn.linear_model import Ridge

reg = Ridge(alpha=.1)

predictors = ["Amb.Temp.-Deg.C","Mod.Temp.-Deg.C","Wind Speed-km/hr","Wind Dir.-Deg.","Humidity-%"]

df = df.sort_index()

train = df.loc[:"2024-12-15 23:45:00"]
test = df.loc["2024-12-16 00:00:00":]

train.tail()


```

```python
print(test['GHI-W/m2'].value_counts())
print(train['GHI-W/m2'].value_counts())
```

```python
reg.fit(train[predictors], train["GHI-W/m2"])
predictions = reg.predict(test[predictors])
from sklearn.metrics import mean_squared_error

mean_squared_error(test["GHI-W/m2"], predictions)
```
